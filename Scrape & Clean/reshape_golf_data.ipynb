```python
# Import dependencies
from bs4 import BeautifulSoup as soup
from splinter import Browser
import pandas as pd
import time
```

```python
# Read in cleaned df for reshaping
cleaned_df = pd.read_csv('cleaned_dataset.csv')
```

```python
# Set index
cleaned_df = cleaned_df.set_index('PLAYER NAME')
```

```python
# Show dataset
cleaned_df.head()
```

```python
# Create subset for each year starting with 2021
df_subset_2021 = cleaned_df[cleaned_df.columns[:23]].copy()

# Store new columns in list
new_columns_2021 = []
for c in df_subset_2021.columns:
    new_columns_2021.append(c.replace("2021_", ''))
```

```python
# Add year columns as a reference
df_subset_2021.columns = new_columns_2021
df_subset_2021['Year'] = '2021'
df_subset_2021
```

```python
# Repeat for other years

# 2020
df_subset_2020 = cleaned_df[cleaned_df.columns[23:46]].copy()

new_columns_2020 = []
for c in df_subset_2020.columns:
    new_columns_2020.append(c.replace("2020_", ''))

df_subset_2020.columns = new_columns_2020
df_subset_2020['Year'] = '2020'
df_subset_2020
```

```python
# Repeat for other years

# 2019
df_subset_2019 = cleaned_df[cleaned_df.columns[46:69]].copy()

new_columns_2019 = []
for c in df_subset_2019.columns:
    new_columns_2019.append(c.replace("2019_", ''))

df_subset_2019.columns = new_columns_2019
df_subset_2019['Year'] = '2019'


# 2018
df_subset_2018 = cleaned_df[cleaned_df.columns[69:92]].copy()

new_columns_2018 = []
for c in df_subset_2018.columns:
    new_columns_2018.append(c.replace("2018_", ''))

df_subset_2018.columns = new_columns_2018
df_subset_2018['Year'] = '2018'


# 2017
df_subset_2017 = cleaned_df[cleaned_df.columns[92:115]].copy()

new_columns_2017 = []
for c in df_subset_2017.columns:
    new_columns_2017.append(c.replace("2017_", ''))

df_subset_2017.columns = new_columns_2017
df_subset_2017['Year'] = '2017'
```

```python
# List of your dataframes
list_dfs = [df_subset_2021, df_subset_2020, df_subset_2019, df_subset_2018, df_subset_2017]  

# Make a new df
reshaped_full_cleaned_dataset = pd.concat(list_dfs)
```

```python
# Display 
reshaped_full_cleaned_dataset.head()
```

```python
reshaped_full_cleaned_dataset.to_csv('reshaped_full_cleaned_dataset.csv')
```

```python
# Drop columns to prep for unsupervised model
reshaped_cleaned_dataset = reshaped_full_cleaned_dataset.reset_index(drop=True, inplace=True)
reshaped_cleaned_dataset = reshaped_full_cleaned_dataset.drop(columns = ['Year'])
reshaped_cleaned_dataset
```

```python
reshaped_cleaned_dataset.to_csv('reshaped_cleaned_dataset.csv', index=False)
```
