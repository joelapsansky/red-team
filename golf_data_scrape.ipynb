```python
# Scrape 2021
for index, url in enumerate(url_list_2021):
    
    start = time.time()
    
    # Print start
    print(f"Start loop {index}")
    
    # Set the executable path and initialize the chrome browser in splinter
    executable_path = {'executable_path': 'chromedriver'}
    browser = Browser('chrome', **executable_path)
    
    # Pause 3 seconds so as not to overwhelm server
    time.sleep(3)
    
    # Visit URL, parse, find table
    browser.visit(url)
    browser.is_element_present_by_css("ul.item_list li.slide", wait_time=1)
    html = browser.html
    golf_soup = soup(html, 'html.parser')
    table = golf_soup.find('table', class_='table-styled')
    
    # Find rows and headings
    table_rows = table.find('tbody').find_all('tr')
    table_headings = table.find('thead').find_all('th')
    
    # Strip column name text and append to dfs
    column_names = [cm.text.strip() for cm in table_headings]
    dataframe_2021[index].append(column_names)
    
    # Strip row text and append to dfs
    for tr in table_rows:
        rows = [tr.text.strip() for tr in tr.find_all('td')]
        dataframe_2021[index].append(rows)
    
    # Convert list to pandas dataframe
    dataframe_2021[index]= pd.DataFrame(dataframe_2021[index])
    
    # Promote headers
    header = dataframe_2021[index].iloc[0]
    dataframe_2021[index] = dataframe_2021[index][1:]
    dataframe_2021[index].columns = header
    
    # Set Index equal to PLAYER NAME
    dataframe_2021[index] = dataframe_2021[index].set_index('PLAYER NAME')
    
    # Quit browser
    browser.quit()
    
    # Print progress
    print(f"End loop {index} | Elapsed loop time: {time.time() - start}")
```
